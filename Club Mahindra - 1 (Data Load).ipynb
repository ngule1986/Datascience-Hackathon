### `LABELS`

```python
from sklearn.preprocessing import LabelEncoder

# COLUMNS FOR LABELLING: ['member_age_buckets', 'cluster_code', 'reservationstatusid_code']

# labels for member_age_buckets
le_age = LabelEncoder()
le_age.fit(train.member_age_buckets.unique())

train['member_age_buckets'] = le_age.transform(train.member_age_buckets)
test['member_age_buckets'] = le_age.transform(test.member_age_buckets)

# labels for cluster_code
le_cluster = LabelEncoder()
le_cluster.fit(train.cluster_code.unique())

train['cluster_code'] = le_cluster.transform(train.cluster_code)
test['cluster_code'] = le_cluster.transform(test.cluster_code)

# labels for reservationstatusid_code
le_reservation = LabelEncoder()
le_reservation.fit(train.reservationstatusid_code.unique())

train['reservationstatusid_code'] = le_reservation.transform(train.reservationstatusid_code)
test['reservationstatusid_code'] = le_reservation.transform(test.reservationstatusid_code)
```

#### `resort id`

```python
# labels for cluster_code
le_resort = LabelEncoder()
le_resort.fit(train.resort_id.unique())

train['resort_id'] = le_resort.transform(train.resort_id)
test['resort_id'] = le_resort.transform(test.resort_id)
```

### `NEW FEATURES`

```python
train['month_of_arrival'] = train.checkin_date.dt.month
test['month_of_arrival'] = test.checkin_date.dt.month
```

```python
train['weekday_of_arrival'] = train.checkin_date.dt.weekday
test['weekday_of_arrival'] = test.checkin_date.dt.weekday
```

```python
train['scheduled_before'] = train.checkin_date -  train.booking_date
test['scheduled_before'] = test.checkin_date - test.booking_date

train['scheduled_before'] = train.scheduled_before.dt.days
test['scheduled_before'] = test.scheduled_before.dt.days
```

```python
train['week_of_year'] = train.checkin_date.dt.week
test['week_of_year'] = test.checkin_date.dt.week
```

```python
train['day_of_month'] = train.checkin_date.dt.day
test['day_of_month'] = test.checkin_date.dt.day
```

```python
train['no_of_nights'] = train.checkout_date - train.checkin_date
test['no_of_nights'] = test.checkout_date - test.checkin_date

train['no_of_nights'] = train.no_of_nights.dt.days
test['no_of_nights'] = test.no_of_nights.dt.days
```

```python
train['roomnights'] = train.no_of_nights
test['roomnights'] = test.no_of_nights
```

```python
train['total_guest'] = train.numberofadults + train.numberofchildren
test['total_guest'] = test.numberofadults + test.numberofchildren
```

```python
train['total_pax'] = train.total_guest
test['total_pax'] = test.total_guest
```

```python
train.drop(['total_guest', 'no_of_nights'], axis=1, inplace=True)
test.drop(['total_guest', 'no_of_nights'], axis=1, inplace=True)
```

```python
from math import ceil

def week_of_month(dt):
    """ Returns the week of the month for the specified date.
    """
    # weekday from monday == 0 ---> sunday == 6
    last_day_of_week_of_month = dt.day + (7 - (1 + dt.weekday()))
    return int(ceil(last_day_of_week_of_month//7.0))
```

```python
train['week_of_month'] = train.checkin_date.apply(week_of_month)
test['week_of_month'] = test.checkin_date.apply(week_of_month)
```
