### `CHECK MISSING %`

```python
null_features = train.isnull().sum()/len(train)

null_features = null_features[null_features > 0]

count_null = len(null_features.index.values)

print(f'There are {count_null} features with missing values.... \n')

# print(null_features.index.values)
```

```python
# Checking features with more than 20% mising values
null_cols = null_features[null_features >= 0.2].index.values

print(f'There are {len(null_cols)} Features with more than 20% missing values.....\n')

print(null_cols)
```

```python
# Features with unique values less than 10 and same unique values in test and train
u_10 = []

for x in null_cols:
    mina = train[x].min()
    maxa = train[x].max()
    count = len(train[x].unique())
    print(f'Feature: {x} | Min Value: {mina} | Max Value: {maxa} | Unique Count: {count}')
    
    if count < 10:
        u_10.append(x)
```

```python
cols_to_drop = []

for x in null_cols:
    if x not in u_10:
        cols_to_drop.append(x)
```

```python
# Drop features with more null values
train.drop(columns = cols_to_drop, inplace=True)
test.drop(columns = cols_to_drop, inplace=True)
```

### `FEATURES WHERE MIN = MAX`

```python
minmax = []

for x in train.drop('Col2', axis=1).columns.values:
    a = train[x].min()
    b = train[x].max()
    
    if a == b:
        print(f'Feature: {x} | Min: {a} | Max: {b}')
        minmax.append(x)
```

```python
# Drop features with more unique values
if len(minmax) > 0:
    train.drop(columns=minmax, axis=1, inplace=True)
    test.drop(columns=minmax, axis=1, inplace=True)
```
