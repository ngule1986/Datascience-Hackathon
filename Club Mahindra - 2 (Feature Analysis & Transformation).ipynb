### `ENVIRONMENT & DATA`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

pd.set_option('display.max_columns', 100)

import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv('train_clean.csv')
test = pd.read_csv('test_clean.csv')
```

```python
train.head()
```

### `DATA TYPES`

```python
cat_columns = ['channel_code', 'main_product_code', 'persontravellingid','resort_region_code', 'resort_type_code', 
               'room_type_booked_code', 'season_holidayed_code', 'state_code_residence', 'state_code_resort',
               'booking_type_code', 'cluster_code', 'reservationstatusid_code', 'resort_id', 'month_of_arrival',
              'weekday_of_arrival', 'week_of_year', 'day_of_month', 'week_of_month', 'member_age_buckets']
```

```python
train[cat_columns] = train[cat_columns].astype('category')
test[cat_columns] = test[cat_columns].astype('category')
```

### `TRANSFORMED FEATURES`

```python
train['roomnights'] = np.log(train.roomnights + 1)
test['roomnights'] = np.log(test.roomnights + 1)
```

```python
train['total_pax'] = np.log(train.total_pax + 1)
test['total_pax'] = np.log(test.total_pax + 1)
```

```python
train['scheduled_before'] = np.log(train.scheduled_before + 1)
test['scheduled_before'] = np.log(test.scheduled_before + 1)
```

```python
df = train.drop('amount_spent_per_room_night_scaled', axis = 1)
df['amount_spent_per_room_night_scaled'] = train.amount_spent_per_room_night_scaled
```

```python
dft = test.drop('reservation_id', axis=1)
dft['reservation_id'] = test.reservation_id
```

### `SAVING FILE`

```python
df.to_csv('train_trans.csv', index=False)
dft.to_csv('test_trans.csv', index=False)
```
